```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(player_df, test_size=0.1, random_state=42)
print("Length of training data:", len(train_set))
print("Length of testing data:", len(test_set))
print("Length of total data:", len(player_df))
```

```python
y_train = train_set['overall']
X_train = train_set.drop(columns = ['overall','attacking_crossing','attacking_finishing','attacking_heading_accuracy','attacking_volleys','skill_dribbling','skill_curve','skill_fk_accuracy','skill_ball_control','movement_acceleration','movement_sprint_speed','movement_agility','movement_balance','power_jumping','power_stamina','power_strength','power_long_shots','mentality_aggression','mentality_interceptions','mentality_positioning','mentality_penalties','defending_marking_awareness','defending_standing_tackle','defending_sliding_tackle','goalkeeping_diving','goalkeeping_handling','goalkeeping_kicking','goalkeeping_positioning','goalkeeping_reflexes'], axis=29)
y_test = test_set['overall']
X_test = test_set.drop(columns = ['overall','attacking_crossing','attacking_finishing','attacking_heading_accuracy','attacking_volleys','skill_dribbling','skill_curve','skill_fk_accuracy','skill_ball_control','movement_acceleration','movement_sprint_speed','movement_agility','movement_balance','power_jumping','power_stamina','power_strength','power_long_shots','mentality_aggression','mentality_interceptions','mentality_positioning','mentality_penalties','defending_marking_awareness','defending_standing_tackle','defending_sliding_tackle','goalkeeping_diving','goalkeeping_handling','goalkeeping_kicking','goalkeeping_positioning','goalkeeping_reflexes'], axis=29)
```

```python
from sklearn.linear_model import LinearRegression
final_model = LinearRegression()
final_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_predictions = final_model.predict(X_train)
lin_mse = mean_squared_error(y_train, y_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(final_model, X_train, y_train,scoring='neg_mean_squared_error', cv=10)
lin_reg_scores = np.sqrt(-scores)
def display_scores(scores):
  print("Scores:", scores)
  print("Mean:", scores.mean())
  print("Standard Deviation:", scores.std())
display_scores(lin_reg_scores)
```

```python
some_data = X_test.iloc[:5]
some_label = y_test.iloc[:5]
print("Predictions:", final_model.predict(some_data))
print("Labels:", list(some_label))
```
